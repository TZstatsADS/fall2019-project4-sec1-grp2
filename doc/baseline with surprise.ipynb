```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from surprise import Reader, Dataset
from surprise.model_selection import train_test_split
from surprise import BaselineOnly, accuracy
from surprise.accuracy import rmse
```

```python
filepath = 'C:/Users/lyc07/Desktop/Columbia/S3/GR5243 Applied Data Science/prj4/'
```

```python
full = pd.read_csv(filepath + 'full.csv')
```

```python
full.head()
```

```python
full.rating.unique()
```

```python
full_data = full[['userId', 'movieId', 'rating']]
```

```python
full_data.head()
```

```python
reader = Reader()
data = Dataset.load_from_df(full_data, reader)
```

```python
train, test = train_test_split(data, test_size = 0.2)
```

```python
print('Using ALS')
bsl_options = {'method': 'als',
               'n_epochs': 5,
               'reg_u': 12,
               'reg_i': 5
               }
algo = BaselineOnly(bsl_options=bsl_options)
```

```python
algo.fit(train)
```

```python
predictions = algo.test(test)
```

```python
predictions
```

```python
rmse(predictions,verbose= True)
```
